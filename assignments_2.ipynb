Qno:8)  Implement a program that checks if a given number is a perfect square without using square
root function. Use binary search approach with while loop.

```python
num = int(input("enter a number: "))
if num < 0:
    print("number is negative")

else:
    low = 0
    high = num
    found = False
    while low <= high:
        mid = (low+high)//2
        sqr = mid * mid
        if sqr == num:
            found = True
            break
        elif sqr > num:
            high = mid - 1
        else:
            low = mid + 1
    if found:
        print("number is perfect sqr")
    else:
        print("number is not perfect sqr")
```

Qno:11) Implement the Sieve of Eratosthenes without using arrays - use nested loops to mark and identify
primes up to n=10000. Count twin primes (primes differing by 2).

```python
n = 10000
num = 2
twin_count = 0
prev_prime = -1

while num <= n:
    
    prime = True
    i = 2
    
    # check prime
    while i*i <= num:
        if num % i == 0:
            prime = False
            break
        i += 1
    
    if prime:
        
        # twin prime check
        if prev_prime != -1 and num - prev_prime == 2:
            twin_count += 1
        
        prev_prime = num
    
    num += 1

print("Twin primes count:", twin_count)
```
