Qno:22) Implement numerical differentiation: approximate f′(x) for f(x) = x^4 − 3x^3 + 2x using central
difference formula with decreasing h values until convergence.

```python
x = float(input("Enter value of x: "))

h = 1.0
previous = 0.0
error = 1.0
while error > 1e-6:
    f_h = (x+h)**4 - 3*(x+h)**3 + 2*(x+h)
    fh = (x-h)**4 - 3*(x-h)**3 + 2*(x-h)
    derivative = (f_h - fh) / (2*h)
    if previous != 0:
        error = abs(derivative - previous)
    previous = derivative
    h = h / 2

print("Approximate derivative =", derivative)
print("Final h value =", h)
print("Final error =", error)
```

```python
import math
a = 0
b = 1
n = 5
R = []

for i in range(n):
    R.append([0.0]*n)
    
for i in range(n):
    N = 2**i
    h = (b - a) / N
    
    total = 0.5 * (1/(1+4*a*a) + 1/(1+4*b*b))
    
    k = 1
    while k < N:
        x = a + k*h
        total += 1/(1+4*x*x)
        k += 1
    
    R[i][0] = h * total
j = 1
while j < n:
    i = j
    while i < n:
        R[i][j] = (4**j * R[i][j-1] - R[i-1][j-1]) / (4**j - 1)
        i += 1
    j += 1
print("Romberg Approximation =", R[n-1][n-1])

exact = 0.5 * math.atan(2)
print("Exact Value =", exact)
print("Difference =", abs(R[n-1][n-1] - exact))
```

Qno:32) Calculate the partial sum of the Basel problem:

$$
\sum_{n=1}^{N} \frac{1}{n^2}
$$

for large N. Estimate the error from the actual value

$$
\frac{\pi^2}{6}.
$$


```python
import math

N = int(input("Enter value of N: "))

S_N = 0.0

for n in range(1, N + 1):
    S_N += 1 / (n * n)

actual_value = (math.pi ** 2) / 6
error = actual_value - S_N

print("Partial Sum S_N =", S_N)
print("Actual Value (pi^2/6) =", actual_value)
print("Estimated Error =", error)
```

Qno:34) Compute the Madhava-Leibniz series for pi:

$$
\frac{\pi}{4} = 1 - \frac{1}{3} + \frac{1}{5} - \frac{1}{7} + \cdots
$$

Apply the **Shanks transformation** to accelerate the convergence of the series.
$$
S_n' = S_{n+2} - \frac{(S_{n+2} - S_{n+1})^2}{S_{n+2} - 2 S_{n+1} + S_n}
$$


```python
import math
n_terms = int(input("Enter number of terms: "))

prev2 = 0.0  
prev1 = 0.0  
current = 0.0  

sum_so_far = 0.0

print("\nPartial sums S_n:")
for k in range(n_terms):
    term = (-1)**k / (2*k + 1)
    sum_so_far += term
    print(f"S_{k+1} = {sum_so_far:.6f}")
    if k >= 2:
        numerator = (current - prev1)**2
        denominator = current - 2*prev1 + prev2
        if denominator != 0:
            S_shanks = current - numerator / denominator
        else:
            S_shanks = current
        print(f"S_{k-1}' (Shanks) = {S_shanks:.6f}")

    prev2 = prev1
    prev1 = current
    current = sum_so_far

pi_over_4 = math.pi / 4
print(f"\nActual π/4 = {pi_over_4:.6f}")
```

Qno:34) Implement the secant method to solve tan(x) = x in the interval [4, 5]. Compare iterations
needed vs bisection method

```python
import math
a = 4.4
b = 4.8
tolerance = 1e-6
max_iter = 100

x0 = a
x1 = b
secant_iter = 0

while secant_iter < max_iter:
    fx0 = math.tan(x0) - x0
    fx1 = math.tan(x1) - x1
    
    if fx1 - fx0 == 0:
        break
    
    x2 = x1 - fx1 * (x1 - x0) / (fx1 - fx0)
    secant_iter += 1
    
    if abs(x2 - x1) < tolerance:
        break
    
    x0 = x1
    x1 = x2

print(f"Secant Method: Root = {x2:.6f} found in {secant_iter} iterations")

a = 4.0
b = 5.0
bisection_iter = 0

while bisection_iter < max_iter:
    c = (a + b) / 2
    fc = math.tan(c) - c
    bisection_iter += 1
    
    if abs(fc) < tolerance or (b - a)/2 < tolerance:
        break
    fa = math.tan(a) - a
    if fa * fc < 0:
        b = c
    else:
        a = c

print(f"Bisection Method: Root = {c:.6f} found in {bisection_iter} iterations")
```

Qno:37) Create a spiral matrix of size n*n filled with numbers 1 to n^2
in spiral order using only loops
and directional conditionals.

```python

n = 3
matrix = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(0)
    matrix.append(row)
top = 0
bottom = n - 1
left = 0
right = n - 1
num = 1  

while num <= n * n:
    for col in range(left, right + 1):
        if num <= n * n:
            matrix[top][col] = num
            num += 1
    top += 1

    for row in range(top, bottom + 1):
        if num <= n * n:
            matrix[row][right] = num
            num += 1
    right -= 1

    for col in range(right, left - 1, -1):
        if num <= n * n:
            matrix[bottom][col] = num
            num += 1
    bottom -= 1

    for row in range(bottom, top - 1, -1):
        if num <= n * n:
            matrix[row][left] = num
            num += 1
    left += 1

for i in range(n):
    for j in range(n):
        print(matrix[i][j], end=" ")
    print()
```

Qno:40) Create a zigzag pattern where numbers increase diagonally. Implement using loops with
direction-changing logic based on row and column parity

```python
n = int(input("Enter size of matrix: "))
matrix = []
for i in range(n):
    row = []
    for j in range(n):
        row.append(0)
    matrix.append(row)

num = 1
for s in range(2*n - 1):  
    if s % 2 == 0:
        
        i = min(s, n-1)
        j = s - i
        while i >= 0 and j < n:
            matrix[i][j] = num
            num += 1
            i -= 1
            j += 1
    else:
        
        j = min(s, n-1)
        i = s - j
        while j >= 0 and i < n:
            matrix[i][j] = num
            num += 1
            i += 1
            j -= 1


for i in range(n):
    for j in range(n):
        print(f"{matrix[i][j]:3}", end=" ")
    print()
```

Qno:43) Create a ”magic square” checker: verify if a given 3Ö3 arrangement (input as 9 separate variables) forms a magic square where all rows, columns, and diagonals sum to the same value

```python
# Input 3x3 matrix
print("Enter 9 numbers for 3x3 square (row by row):")
matrix = []
for i in range(3):
    row = []
    for j in range(3):
        row.append(int(input(f"({i+1},{j+1}): ")))
    matrix.append(row)

# Compute sum of first row manually
magic_sum = 0
for num in matrix[0]:
    magic_sum += num

is_magic = True

# Check rows and columns
for i in range(3):
    row_sum = 0
    col_sum = 0
    for j in range(3):
        row_sum += matrix[i][j]
        col_sum += matrix[j][i]
    if row_sum != magic_sum or col_sum != magic_sum:
        is_magic = False

# Check diagonals
diag1 = 0
diag2 = 0
for i in range(3):
    diag1 += matrix[i][i]
    diag2 += matrix[i][2-i]
if diag1 != magic_sum or diag2 != magic_sum:
    is_magic = False

print("MAGIC SQUARE" if is_magic else "NOT a magic square")
```
