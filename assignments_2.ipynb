Write a program to determine if a number is prime. If prime, find the next prime number.
If composite, find all its prime factors using trial division with optimization (only check up to
sqrt(n)).

```python
n = int(input("Enter number: "))

i = 2
prime = True

if n <= 1:
    print("negative or zero")

else:
    while i * i <= n:
        if n % i == 0:
            prime = False
            break
        i += 1
    
    
    if prime:
        print("Prime")
    
        next_num = n + 1
        while True:
            i = 2
            while i * i <= next_num and next_num % i != 0:
                i += 1
            if i * i > next_num:
                print("Next prime:", next_num)
                break
            next_num += 1
    
    else:
        print("Composite")
        temp = n
        i = 2
        while i * i <= temp:
            while temp > i and temp % i == 0:
                print(i)
                temp //= i
            i += 1
        if temp > 1:
            print(temp)
```

Implement a program that checks if a given number is a perfect square without using square
root function. Use binary search approach with while loop.

```python
num = int(input("enter a number: "))
if num < 0:
    print("number is negative")

else:
    low = 0
    high = num
    found = False
    while low <= high:
        mid = (low+high)//2
        sqr = mid * mid
        if sqr == num:
            found = True
            break
        elif sqr > num:
            high = mid - 1
        else:
            low = mid + 1
    if found:
        print("number is perfect sqr")
    else:
        print("number is not perfect sqr")
```
